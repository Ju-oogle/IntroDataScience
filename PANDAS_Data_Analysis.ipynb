## Loc & Iloc Attributes

When the index is int type we use iloc
When the index is string type, loc

```python
new_users = {  # Dictionary
    'username': ['user1','user2','user3', 'user4', 'user5'],
    'email': ['user1@example.com','user2@example.com','user3@example.com', 'user4@example.com', 'user4@example.com'],
    'age': [17,35,21, 40, 50],
    'status': [True,False, False, True, False]
}
```

```python
# Checking the indexes type
new_usersdf.index
```

```python
new_usersdf.loc['a']  # Getting the data from row 'a'
```

```python
new_usersdf.loc['e'] # Getting the data from row 'e'
```

```python
new_usersdf.loc['a':'c'] # Getting a sub-dataframe from rows 'a' to 'c'
```

## Searching data by Ordering it

```python
users_df
```

```python
# Get the youngest user from Canada
users_df[ users_df['pais'] == 'Canada']  # DF with Canadian users
```

```python
users_df[ users_df['pais'] == 'Canada'].sort_values('edad').head(1) 
# DF with Canadian users with sorted ages from youngest to oldest, selecting
#the first row
```

```python
# Get the 5 oldest users from Germany
users_df[ users_df['pais'] == 'Germany']  # DF with Canadian users
```

```python
users_df[ users_df['pais'] == 'Canada'].sort_values('edad').tail(5)   
# Get the 5 oldest users from Germany. From the youngest to the oldest.
```

```python
users_df[ users_df['pais'] == 'Canada'].sort_values('edad', ascending = False).head(5) 
# Get the 5 oldest users from Germany. From the oldest to the youngest.
```

```python
# Get the 'name' of all the users older than 30 years old from 'Canada', 'Germany' and 'France'
```

```python
users_df[ users_df['edad'] > 30]  # Getting the users older than 30
```

```python
users_df[ (users_df['edad'] > 30) & (users_df['pais'] == 'Canada') | (users_df['pais'] == 'France') | (users_df['pais'] == 'Germany')  ] 
# To add the different conditions use the assignment operators: & | and () on every condition
```

```python
countries = ['Canada','France','Germany']  # List of the countries to find 
```

```python
users_df[ (users_df['edad'] > 30) & (users_df['pais'].isin(countries))]  # Usage of isin() function
```

## Grouping Data



```python
# TASK:  Find the number of women and males od the DS
users_df.groupby('genero')  # creates an object DF type
```

```python
users_df.groupby('genero')['genero'] # creates an object Series type by specifying the
# key used to group the data
```

```python
users_df.groupby('genero')['genero'].count()  # returns the number of males and females
```

```python
# TASK: Find the country with more women
users_df[ users_df['genero'] == 'female']
```

```python
users_df[ users_df['genero'] == 'female'].groupby('pais') # creates an object DF type
```

```python
users_df[ users_df['genero'] == 'female'].groupby('pais')['pais'].count() # creates an object Series type by specifying the
# key used to group the data this way it can access to .count() function
```

```python
users_df[ users_df['genero'] == 'female'].groupby('pais')['pais'].count().sort_values()
# It sorts the results by ascending values
```

```python
users_df[ users_df['genero'] == 'female'].groupby('pais')['pais'].count().sort_values(ascending=False)
# it orders the values in a descending mode
```

```python
users_df[ users_df['genero'] == 'female'].groupby('pais')['pais'].count().sort_values(ascending=False).head(1)
# It selects the first result, the one that answers the task
```

```python
# Get the names of all users from 'New Zealand'	
same_users = pd.read_csv('users.csv', index_col='id')
same_users
```

```python
same_users = same_users.dropna()  # Deletes the NaN rows if there are
```

```python
same_users['pais'] == 'Canada'   # Returns bool result for the series' conditional
```

```python
same_users[same_users['pais'] == 'Canada']  # We include the series' conditional to the DF to get the results
```

```python
same_users[same_users['pais'] == 'Canada']['nombre']  # Returns the name of the Canadian users
```

```python
# TASK: Get the 'name' and 'email' of all users older than '50' years old
same_users[same_users['edad'] > 50]  # Gets all the users older than 50 years old
```

```python
same_users[same_users['edad'] > 50][['nombre','email']]  # Returns the 'name'
# and 'email' columns of all the users older than 50 years old.
```

```python
# TASK: Get the mean of all 'female' users older than '30' years
same_users[(same_users['genero'] == 'female') & (same_users['edad'] > 30)]  
# Returns all the females older than 30 years old
```

```python
# Get the 'age' column from the DF of all the females older than 30 years old
same_users[(same_users['genero'] == 'female') & (same_users['edad'] > 30)]['edad'] 
```

```python
# Applies the mean() function to the result of all the females older than 30 years old
same_users[(same_users['genero'] == 'female') & (same_users['edad'] > 30)]['edad'].mean() 
```
